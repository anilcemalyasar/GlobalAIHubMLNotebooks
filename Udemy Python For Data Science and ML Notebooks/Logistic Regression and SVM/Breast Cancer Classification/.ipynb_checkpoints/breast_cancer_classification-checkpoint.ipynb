```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("breast-cancer.csv")
```

```python
# As we can see diagnosis(label) is our target in this classification problem
# and all the other features are numerical
data.head(5)
```

```python
data.info()
```

```python
data.describe().transpose()
```

```python
# it means dataset contains 569 instances with 31 features
data.shape
```

```python
# For binary classification we have to transform text-based classes into numerical binary classes
data["diagnosis"].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["diagnosis"] = encoder.fit_transform(data["diagnosis"].values)
```

```python
# we’re trying to predict if the tumor is malignant (1) or benign (0). 
# data["diagnosis"].nunique() --> 2 because it contains only 1 and 0
data["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.3)
```

```python
X_train = train.drop("diagnosis", axis=1)
y_train = train.loc[:, "diagnosis"]

X_test = test.drop("diagnosis", axis=1)
y_test = test.loc[:, "diagnosis"]
```

```python
# There is no label(target) column any more, it contains only the features
X_train
```

```python
# There are not any features, only labels which are 0 and 1
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Creating a logistic regression model
model = LogisticRegression()
```

```python
# Training the data with X_train and y_train
# Supervised learning because labels are given 
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
# As we can see model has decided whether the tumor is malignant (1) or benign (0). 
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# 117 True Negatives  are the instances which were predicted as negatives and their ground truth was also negative.
# 52 True Positives are the instances which were predicted as positives and their ground truth was also positives.
# 2 False Negatives are the instances which were predicted as malignant(1) but their ground truth was benign(0)
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
